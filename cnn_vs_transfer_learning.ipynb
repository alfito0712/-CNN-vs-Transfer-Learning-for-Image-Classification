```
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras import layers,models
print(tf.__version__)
```

## CNN From Scratch

```
# Load CIFAR10 and filter cats/dogs
(x_train,y_train),(x_test,y_test)=tf.keras.datasets.cifar10.load_data()
mask=((y_train==3)|(y_train==5)).flatten()
x_train=x_train[mask]/255.0
y_train=(y_train[mask]==5).astype(int)
mask=((y_test==3)|(y_test==5)).flatten()
x_test=x_test[mask]/255.0
y_test=(y_test[mask]==5).astype(int)
model=models.Sequential([
layers.Input((32,32,3)),
layers.Conv2D(32,3,activation='relu'),
layers.MaxPooling2D(),
layers.Conv2D(64,3,activation='relu'),
layers.MaxPooling2D(),
layers.Flatten(),
layers.Dense(128,activation='relu'),
layers.Dropout(0.5),
layers.Dense(1,activation='sigmoid')
])
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
# history=model.fit(x_train,y_train,epochs=10,validation_split=0.15)
```

## Transfer Learning

```
# Update DATA_DIR to your Kaggle Cats vs Dogs dataset
DATA_DIR='/kaggle/input/your-dataset'
img_size=(224,224)
train_ds=tf.keras.utils.image_dataset_from_directory(DATA_DIR,validation_split=0.15,subset='training',seed=42,image_size=img_size,batch_size=32)
val_ds=tf.keras.utils.image_dataset_from_directory(DATA_DIR,validation_split=0.15,subset='validation',seed=42,image_size=img_size,batch_size=32)
base=tf.keras.applications.MobileNetV2(include_top=False,input_shape=(224,224,3),weights='imagenet')
base.trainable=False
inputs=tf.keras.Input((224,224,3))
x=tf.keras.applications.mobilenet_v2.preprocess_input(inputs)
x=base(x,training=False)
x=layers.GlobalAveragePooling2D()(x)
x=layers.Dense(128,activation='relu')(x)
outputs=layers.Dense(1,activation='sigmoid')(x)
tl=tf.keras.Model(inputs,outputs)
tl.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
# history_tl=tl.fit(train_ds,validation_data=val_ds,epochs=5)
```
